```python
# h
import matplotlib.pyplot as plt

# Sample data: Hero roles and their counts
heroes_data = {
    'Tank': 10,
    'Support': 15,
    'DPS': 25,
    'Healer': 5
}

# Calculate total number of heroes
total_heroes = sum(heroes_data.values())

# Calculate percentage for each role
roles_percentages = {role: (count / total_heroes) * 100 for role, count in heroes_data.items()}

# Plotting the data
plt.figure(figsize=(8, 6))

# Pie chart
plt.pie(roles_percentages.values(), labels=roles_percentages.keys(), autopct='%1.1f%%', startangle=140)
plt.title('Percentage of Heroes Based on Their Roles')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the plot
plt.show()
```

7. I have a character with this specification:
defense_overall 6.0<br>
offense_overall 7.0<br>
skill_effect_overall 4.0<br>
difficulty_overall 3.0<br>
movement_spd 250<br>
magic_defense 9<br>
mana 380<br>
hp_regen 37<br>
physical_atk 119<br>
physical_defense 24<br>
hp 2689<br>
attack_speed 0.75<br>
mana_regen 16.0<br>

Please predict its role using DecisionTree classification algorithm <br>

```python
# 7
# import library

# Encode the 'role' column

# Separate features and target variable

# Split the data into training and testing sets

# Create a Decision Tree Classifier

# Train the model

# Make predictions on the test set

# Evaluate the model

# Predict the role for the given character

# Get the predicted role

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Step 1: Load the dataset
df = pd.read_csv("mlbb_hero.csv")

# Display the first few rows to understand the data structure
print(df.head())

# Check the actual column names
print("Columns in the dataset:", df.columns)

# Step 2: Encode the 'Role' column (target variable)
# Make sure the 'Role' column exists and is named correctly
df.columns = df.columns.str.strip()  # Remove any leading/trailing whitespaces
if 'role' not in df.columns:
    raise KeyError("The 'Role' column was not found in the dataset. Please check your CSV file.")

# Encode the 'Role' column
label_encoder = LabelEncoder()
df['role'] = label_encoder.fit_transform(df['role'])

# Step 3: Separate features (X) and target variable (y)
features = ['defense_overall', 'offense_overall', 'skill_effect_overall', 'difficulty_overall',
            'movement_spd', 'magic_defense', 'mana', 'hp_regen', 'physical_atk', 
            'physical_defense', 'hp', 'attack_speed', 'mana_regen']

# Ensure that all feature columns are present in the dataset
missing_features = [feature for feature in features if feature not in df.columns]
if missing_features:
    raise KeyError(f"The following features are missing from the dataset: {missing_features}")

X = df[features]
y = df['role']

# Step 4: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 5: Create a Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)

# Step 6: Train the model
clf.fit(X_train, y_train)

# Step 7: Make predictions on the test set
y_pred = clf.predict(X_test)

# Step 8: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Step 9: Predict the role for the given character
# Convert new_character to a DataFrame to match the feature names
new_character = pd.DataFrame([[6.0, 7.0, 4.0, 3.0, 250, 9, 380, 37, 119, 24, 2689, 0.75, 16.0]], columns=features)

predicted_role = clf.predict(new_character)

# Get the actual role name
predicted_role_name = label_encoder.inverse_transform(predicted_role)
print(f"Predicted role for the character: {predicted_role_name[0]}")

```
